### train Autoregressive model | univariate - (1)

Assess stationarity

```python
from statsmodels.tsa.stattools import adfuller
def stationarity_check(timeseries,rolling_window=3):
    mpl.rc('figure', figsize=(5.5, 3.5))
    #calcualte rolling stats (mean/std deviation)
    rolmean = pd.Series(timeseries).rolling(window=rolling_window).mean().dropna()
    rolstd = pd.Series(timeseries).rolling(window=rolling_window).std().dropna()

    #plot rolling stats:
    orig = plt.plot(timeseries, color='green',label='data')
    mean = plt.plot(rolmean.values, color='purple', label='rolling_mean')
    std = plt.plot(rolstd.values, color='#295673', label = 'rolling_std',alpha=0.7)
    plt.legend(loc='best')
    plt.title('raw data + rolling stats')
    plt.show(block=False)
    
    print('dickey-fuller test:')
    print('=-=--==--=-=-=-==-=')
    dickey_fuller_results = adfuller(timeseries, autolag='AIC')  #returns tuple
    dfoutput = pd.Series(dickey_fuller_results[0:4], index=['test_stat','p_val','lag_num','n_obs'])  #[0:4] excludes 1,5,10% critical vals
    for key,value in dickey_fuller_results[4].items():
        dfoutput['crit vals ({0})'.format(key)] = value
    print(dfoutput)
```

```python
timeseries = df[target_column].values
stationarity_check(timeseries,rolling_window=3)
```

### Basic Auto-Regressive model

#### train model on raw data

```python
from statsmodels.tsa.ar_model import AR
```

```python
# select column
ar_df = pd.DataFrame()
ar_df[target_column] = df[target_column]
```

```python
# rename column
base_col = 't'
ar_df.rename(columns={target_column: base_col}, inplace=True)
```

```python
# partition data into train/test sets
train_cut_pct = 0.85
cut_period = len(ar_df) - int(len(ar_df)*train_cut_pct)
# print(cut_period)
```

```python
X = ar_df.values
train, test = X[1:len(X)-cut_period], X[len(X)-cut_period:]
print(len(X))
print(len(train))
print(len(test))
```

```python
# ar_model = AR(train, freq='W')
ar_model = AR(train)
ar_model_fit = ar_model.fit()
```

```python
# evalulation stats
print('lag: {0}'.format(ar_model_fit.k_ar))
print('bic: {0}'.format(ar_model_fit.bic))
print('aic: {0}'.format(ar_model_fit.aic))
```

```python
# explore calculated coefficients
print('coefficients: {0}'.format(ar_model_fit.params))
```

```python
# generate predictions
predictions = ar_model_fit.predict(start=len(train), end=len(train)+len(test)-1, dynamic=False)
```

```python
# summary estimate of model performance
error = mean_squared_error(test, predictions)
print('MSE (test): {0}'.format(error))
```

```python
# plot predictions vs expected
plt.plot(test,color='green',lw=2)
plt.plot(predictions, color='purple',lw=2.5,alpha=0.85)
plt.grid(True)
plt.legend(['actual','pred'], loc='best')
plt.title('fail')
plt.show()
```

### Enforce stationarity, retrain model after differencing data

```python
# use period over period difference as input feed (to achieve stationarity)
lag_n = 1
print(len(ar_df))
ar_df['diff'] = (ar_df[base_col] - ar_df[base_col].shift(lag_n))
ar_df = ar_df.dropna().copy()
print(len(ar_df))
```

```python
ar_df.head(5)
```

```python
# partition data into train/test sets
X = ar_df['diff'].values
train, test = X[1:len(X)-cut_period], X[len(X)-cut_period:]
print(len(X))
print(len(train))
print(len(test))
```

```python
# instantiate model / fit
ar_model = AR(train) # freq='W'
ar_model_fit = ar_model.fit()
```

```python
# summary stats
print('lag: {0}'.format(ar_model_fit.k_ar))
print('bic: {0}'.format(ar_model_fit.bic))
print('aic: {0}'.format(ar_model_fit.aic))
```

```python
# review coefficients
print('coefficients: {0}'.format(ar_model_fit.params))
```

```python
# generate predictions
predictions = ar_model_fit.predict(start=len(train), end=len(train)+len(test)-1, dynamic=False)
```

```python
error = mean_squared_error(test, predictions)
print('MSE (test): {0}'.format(error))
```

```python
# plot predictions vs expected
mpl.rc('figure', figsize=(6, 5))
plt.plot(test,color='green',lw=2)
plt.plot(predictions, color='purple',lw=2.5,alpha=0.85)
plt.grid(True)
plt.legend(['actual','pred'], loc='best')
plt.show()
```

#### test ARIMA approach to AR

```python
# set underlying data series
timeseries = df[target_column].values
```

```python
from statsmodels.tsa.arima_model import ARIMA
# train model with d = 1 (integrate), no MA
model = ARIMA(timeseries, order=(1, 1, 0)) 
model_fit = model.fit(disp=0)
```

```python
print(model_fit.summary())
```

```python
mpl.rc('figure', figsize=(6, 3))
residuals = pd.DataFrame(model_fit.resid)
residuals.plot(color='purple')
plt.show()
```

```python
residuals.plot(kind='kde')
plt.show()
```

```python
print(residuals.describe())
```

### convert (diff) predictions back into yields

```python
# add predictions back to df object [accommodate specified lag]
ar_df['pred'] = model_fit.fittedvalues[(lag_n-1):]
```

```python
ar_df.head(5)
```

```python
# plot actual vs predictions
# should not exhibit any linear pattern when model 'fits' well
# -------------------------
mpl.rc('figure', figsize=(3, 3))
xxx = ar_df['diff'].values
yyy = ar_df['pred'].values
plt.scatter(xxx, yyy, s=80, facecolors='none', edgecolors='purple',alpha=0.3)
plt.show()
```

```python
# create t-1 lag
ar_df['t-1'] = ar_df[base_col].shift(1)
```

```python
# lag predicted diff
ar_df['tmp'] = ar_df['pred'].shift(-1).values
```

```python
# compute predicted yield based on predicted diff
ar_df['t+1_pred'] = ar_df['tmp'] + ar_df[base_col]
ar_df['t+1_pred'] = ar_df['t+1_pred'].shift(1)
```

```python
# reorder columns | remove tmp data
ar_df = ar_df[['t-1',base_col,'t+1_pred','pred','diff']]
```

```python
# drop NaNs
ar_df.dropna(inplace=True)
```

```python
ar_df['resid'] = ar_df['t+1_pred'] - ar_df[base_col]
```

```python
ar_df.head(5)
```

```python
# quick sanity check that residuals normally dist
mpl.rc('figure', figsize=(4, 2))
ar_df['resid'].plot(kind='kde')
plt.axvline(x=0,lw=1,linestyle='--',color='k')
plt.show()
```

## Model summary

```python
# plot predictions vs expected
mpl.rc('figure', figsize=(6, 5))
plt.plot(ar_df[base_col].values,color='green',lw=2)
plt.plot(ar_df['t+1_pred'].values, color='purple',lw=4,alpha=0.4)
plt.grid(True)
plt.legend(['actual','pred'], loc='best')
plt.show()
```

```python
# Zooming in for better visual asssessment of residuals
# plot predictions vs expected
mpl.rc('figure', figsize=(6, 5))
plt.plot(ar_df[base_col].values,color='green',lw=2)
plt.plot(ar_df['t+1_pred'].values, color='purple',lw=2,alpha=0.9)
plt.grid(True)
plt.legend(['actual','pred'], loc='best')
plt.xlim(500,550)
plt.show()
```

```python
# MSE calc:
ljust_parma=14
print('MSE')
print('------------')
print('manual calc:'.ljust(ljust_parma),  str(np.sum((ar_df['resid'])**2) / len(ar_df))  )
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(ar_df['diff'].values, ar_df['pred'].values)  
print('sklearn:'.ljust(ljust_parma),  str(mse)  )
```

#### additional residual exploration

```python
# residual plot:
mpl.rc('figure', figsize=(4, 4))

# create percentage diff (residual / period diff)
ar_df['resid_pct'] = ar_df['resid'] / ar_df[base_col]

# scatter plot diff vs. pct miss
x_input = 'diff'
y_input = 'resid_pct'
xxx = ar_df[x_input].values[1:]
yyy = ar_df[y_input].values[1:]
plt.scatter(xxx, yyy, s=80, facecolors='none', edgecolors='purple')
plt.xlabel('{0}'.format(x_input))
plt.ylabel('{0}'.format(y_input))
plt.axvline(x=0,ymin=0.2,ymax=0.8,lw=1,linestyle='--',color='k')
plt.axhline(y=0,xmin=0.2,xmax=0.8,lw=1,linestyle='--',color='k')

plt.xlim(min(ar_df[x_input].min(),ar_df[y_input].min()),max(ar_df[x_input].max(),ar_df[y_input].max()))
plt.ylim(min(ar_df[x_input].min(),ar_df[y_input].min()),max(ar_df[x_input].max(),ar_df[y_input].max()))
plt.show()
```
